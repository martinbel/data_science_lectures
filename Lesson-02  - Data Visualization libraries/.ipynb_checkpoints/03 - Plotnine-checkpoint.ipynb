## 1. Scatter Plots - geom_point()

Docs: https://plotnine.readthedocs.io/en/stable/generated/plotnine.geoms.geom_point.html#plotnine.geoms.geom_point

```python
# aes in ggplot() call works
(ggplot(mpg, aes(x='displ', y='hwy')) +
 geom_point()
)
```

```python
# with a one liner we don't need the extra 
ggplot(mpg, aes(x='displ', y='hwy')) + geom_point()
```

```python
# aes in geom_point() call works also
(ggplot(mpg) +
 geom_point(aes(x='displ', y='hwy'))
)
```

```python
# cyl is mapped as numeric
ggplot(mpg) + geom_point(aes(x='displ', y='hwy', color='cyl'))
```

```python
# Convert an integer variable to categorical (a factor in R parlance)
p = ggplot(mpg, aes(x='displ', y='hwy'))
p + geom_point(aes(color='factor(cyl)'))
```

```python
p + geom_point(aes(color='cty')) + scale_color_gradient(low='blue', high='red')
```

## 2. Scatter Plots - geom_boxplot()

Docs: https://plotnine.readthedocs.io/en/stable/generated/plotnine.geoms.geom_boxplot.html#plotnine.geoms.geom_boxplot

```python
for col in ['cyl', 'year']:
    mpg[col] = mpg[col].astype("category")
```

```python
mpg.head()
```

```python
mpg.dtypes
```

```python
ggplot(mpg) + geom_boxplot(aes(x='cyl', y='hwy'))
```

```python
# Use two categorical variables by one numeric variable
ggplot(mpg) + geom_boxplot(aes(x='cyl', y='displ', fill='year'))
```

```python
economics.head()
```

```python
economics.dtypes
```

```python
economics['year'] = pd.DatetimeIndex(economics.date).year.astype('category')
```

```python
economics['unemployment_rate'] = (economics['unemploy'] / economics['pop']) * 100
```

```python
(ggplot(economics, aes(x='year', y='unemployment_rate')) + 
 geom_boxplot() + 
 theme(axis_text_x = element_text(angle=90, hjust=1))
)
```

```python
# change x-axis breaks formatting
def custom_date_format1(breaks):
    return [x.year if x.month==1 and x.day==1 else "" for x in breaks]


(ggplot(economics) + 
 geom_point(aes('date', 'psavert')) + 
 scale_x_datetime(breaks=date_breaks('5 years'), labels=custom_date_format1) + 
 labs(y='Personal saving rate', x='Date')
)
```

## 3. Facets

- Facet wrap: https://plotnine.readthedocs.io/en/stable/generated/plotnine.facets.facet_wrap.html#plotnine.facets.facet_wrap
- Facet grid: https://plotnine.readthedocs.io/en/stable/generated/plotnine.facets.facet_grid.html#plotnine.facets.facet_grid

```python
# One scale for all sub-plots, default
(ggplot(mpg) + 
 geom_histogram(aes(x='hwy')) + 
 facet_wrap("~ class"))
```

```python
# Making scales of each plot independent
(ggplot(mpg) + 
 geom_histogram(aes(x='hwy')) + 
 facet_wrap("~ class", scales='free_y'))
```

```python
# Making scales of each plot independent
(ggplot(mpg) + 
 geom_histogram(aes(x='hwy')) + 
 facet_wrap("~ cyl"))
```

```python
(ggplot(mpg, aes(x='displ', y='hwy')) + 
 geom_point() + 
 facet_wrap("~ class", scales='free_y'))
```

```python
# One facet grid example. 
# Facet grid is mostly used for plotting 2 categorical variables whereas facet_wrap is used for plotting 1 categorical variable.

(ggplot(mpg, aes(x='displ')) + 
 geom_histogram() + 
 facet_grid("cyl ~ year"))
```

```python
# You can add models to point plots. 
# I often prefer rlm that is a robust linear model to the basic lm method (OLS).
(ggplot(mtcars, aes('wt', 'mpg', color='factor(gear)'))
 + geom_point()
 + geom_smooth(method='rlm')
 + facet_wrap('~gear'))
```
