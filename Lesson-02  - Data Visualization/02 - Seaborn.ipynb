```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
sns.set(style="darkgrid")
```

```python
cars = pd.read_csv("data/mtcars.csv")
diamonds = pd.read_csv("data/diamonds.csv")
```

```python
cars.head()
```

```python
### Scatter plots
```

```python
# Grouping by a categorical variable
sns.relplot(x="displ", y="hwy", hue="class", data=cars)
```

```python
# Grouping by a categorical variable that is an integer
sns.relplot(x="displ", y="hwy", hue="cyl", palette="Blues", data=cars)
```

```python
# Using size, often not the clearest way to visualize relations
sns.relplot(x="displ", y="hwy", size="cyl", data=cars)
```

```python
# Using shape (or style) instead of size and colour. This is often useless
sns.relplot(x="displ", y="hwy", style="cyl", data=cars)
```

```python
### Facets
# Using the high level function relplot
sns.relplot(x="displ", y="hwy", col="class", data=cars, col_wrap=3) # hue="cyl",
```

```python
# The same using FacetGrid
g = sns.FacetGrid(cars, col="class", col_wrap=3)   # hue="cyl",
g.map(plt.scatter, "displ", "hwy", edgecolor="w").add_legend()
```

```python
# Define col, row and other mappings
g = sns.FacetGrid(cars, col="class", col_wrap=3)

# map a function to the plot such as: plt.hist, plt.scatter, etc.
g.map(plt.hist, "hwy", edgecolor="w").add_legend()
```

```python
g = sns.FacetGrid(cars, row="class", col="cyl")
g.map(plt.scatter, "displ", "hwy", edgecolor="w")
```
